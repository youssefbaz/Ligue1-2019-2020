```python
data=pd.read_csv('data_2020_2021.csv')
data.head(10)
```

```python
#list all the games of the first jounrey 

data.loc[data['journee'] == '1. Journée']
```

```python
# list all the games of PSG where he was playing at home 
#psg_games= data.loc[data['home']== 'Paris Saint-Germain'] + data.loc[data['away']=='Paris Saint-Germain']
#away_psg=data.loc[data['away']== 'Paris Saint-Germain']
#home_psg

df2 = data.loc[data['home']== 'Paris Saint-Germain'] 
df3=data.loc[data['away']=='Paris Saint-Germain']
psg_games=pd.concat([df2, df3], axis=0)
psg_games
```

```python
#we ordered the list of games of Paris Saint Germain by combining home and away games 

psg_games['journey']=psg_games['journee'].apply(lambda x: x.zfill(3)).str[0:2]
psg_games['journey']=[i.replace(".","")for i in psg_games['journey']]
psg_games['journey']=psg_games['journey'].apply(lambda x: x.zfill(2))
psg_games=psg_games.sort_values(by=['journey'])
psg_games
```
